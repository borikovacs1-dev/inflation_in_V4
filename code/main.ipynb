## What can we say about the evolution of inflation in V4 countries in the last 10 years
### Inflation
* Inflation was realtively constant between 2017 and 2021 in all countries
* From the second half of 2021 we see a large inflation spike
* The average inflation peaked at the beginning of 2023 almost reaching 20%
* Hungary experienced the largest inflation spike where inflation peaked at more that 25%

### Core inflation
* Core inflation should be less volatile than inflation as it excludes those products that have highly volatile prices, this cannot be directly seen on the graphs.
* Core inflation closely follows inflation
* In the last few months core inflation in V4 countries diverged
